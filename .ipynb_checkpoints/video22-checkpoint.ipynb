## PCA and its effect on supervised learning
This exercise will demonstrate how principal component analysis (PCA) can allow for successful supervised learning on datasets that have been dramatically reduced in size. I will use the breast cancer dataset from the Scikit-Learn library to first measure the difference in performance on principal-component-fitted models to their full-data counterparts, and fit an effective classifier using reduced data.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn import svm
```

```python
# Read in cancer data
cancer_data = load_breast_cancer()
cancer_data.data.shape

# Standardize data
scaler = StandardScaler().fit(cancer_data.data)
data_scaled = scaler.transform(cancer_data.data)
```

```python
# Create test/train split
X_train_orig, X_test_orig, y_train, y_test = train_test_split(
    data_scaled,
    cancer_data.target,
    stratify = cancer_data.target,
    shuffle = True,
    random_state = 314
)
```

```python
# Perform PCA with 5 components
pca = PCA(n_components = 5)
pca.fit(data_scaled)
data_reduced = pca.transform(data_scaled)
print("Original shape: {}".format(str(data_scaled.shape)))
print("Reduced shape: {}".format(str(data_reduced.shape)))
```

```python
# Convert reduced data to dataframe and add target variable
df = pd.DataFrame(
    data_reduced,
    columns=['PC1','PC2','PC3','PC4','PC5']
)
df['target'] = cancer_data.target
```

```python
# Plot the first two components
df.plot(
    kind = "scatter", 
    x = "PC1", 
    y = "PC2", 
    alpha = 0.4,
    label = "malignant",
    color = np.where(df['target'] == 0, 'r', 'b')
)
```

```python
# Plot the next two components
df.plot(
    kind = "scatter", 
    x = "PC3", 
    y = "PC4", 
    alpha = 0.4,
    label = "malignant",
    color = np.where(df['target'] == 0, 'r', 'b')
)
```

The first two principal components appear to work effectively to partition the data. The next two do not do as well, but may still prove useful in classifying edge-cases.

```python
# Create test/train split on PCA data
X_train_pca, X_test_pca = train_test_split(
    df.drop('target',axis = 1),
    stratify = df['target'],
    shuffle = True,
    random_state = 314
)
```

```python
# Measure performance on full, standardized dataset
rf_orig = RandomForestClassifier(oob_score = True, n_estimators = 100, random_state = 314)
rf_orig = rf_orig.fit(X_train_orig, y_train)
print("Training set score: {:.3f}".format(rf_orig.score(X_train_orig, y_train)))
print("Test set score: {:.3f}".format(rf_orig.score(X_test_orig, y_test)))
print("OOB score: {:.3f}".format(rf_orig.oob_score_))
```

```python
# Fit random forest on PCA data
rf_pca = RandomForestClassifier(oob_score = True, n_estimators = 100, random_state = 314)
rf_pca = rf_pca.fit(X_train_pca, y_train)
print("Training set score: {:.3f}".format(rf_pca.score(X_train_pca, y_train)))
print("Test set score: {:.3f}".format(rf_pca.score(X_test_pca, y_test)))
print("OOB score: {:.3f}".format(rf_pca.oob_score_) )
```

```python
# Confusion matrix of original RF predictions
pred_orig = rf_orig.predict(X_test_orig)
confusion = confusion_matrix(y_test, pred_orig)
print("Confusion matrix:\n{}".format(confusion))
```

```python
# Confusion matrix of PCA RF predictions
pred_pca = rf_pca.predict(X_test_pca)
confusion = confusion_matrix(y_test, pred_pca)
print("Confusion matrix:\n{}".format(confusion))
```

As you can see from the respective accuracy scores for each model, fitting a random forest on the principal components worked almost as effectively as fitting the model on the entire dataset. While the difference in fitting time is negligible in this example, this technique will be a life-saver when working with higher-dimensional data.

```python
# Fit SVM on PCA data
svm_model_pca = svm.LinearSVC(
    C = 50.0, 
    random_state = 314, 
    max_iter = 70000
).fit(X_train_pca, y_train)

# Make prediction anc check accuracy
pred_svm_pca = svm_model_pca.predict(X_test_pca)
print(f1_score(y_test, pred_svm_pca))

pred_pca = svm_model_pca.predict(X_test_pca)
confusion_matrix(y_test, pred_pca)
```

```python
# Fit SVM on PCA data with adjusted C parameter
svm_model = svm.LinearSVC(
    C = 30.0, 
    random_state = 314, 
    max_iter = 70000
)
svm_model.fit(X_train_pca, y_train)
print("Train set accuracy: {:.2f}".format(svm_model.score(X_train_pca, y_train)))
print("Test set accuracy: {:.2f}".format(svm_model.score(X_test_pca, y_test)))

pred_svm = svm_model.predict(X_test_pca)
confusion = confusion_matrix(y_test, pred_svm)
print("Confusion matrix:\n{}".format(confusion))
```

```python
# Create function to plot model accuracy with various PC numbers
def plot_pca_accuracy(model, train, test, ytrain, ytest):
    components = []
    scores = []
    for c in range(5, 25):
        pca = PCA(n_components = c, whiten = True, random_state = 314).fit(train)
        Xtrain = pca.transform(train)
        Xtest = pca.transform(test)
        model.fit(Xtrain, ytrain)
        score = model.score(Xtest, ytest)
        scores.append(score)
        components.append(c)
    plt.plot(components,scores,label = "train accuracy")
    plt.ylabel("Accuracy")
    plt.xlabel("components")
```

```python
# Plot accuracy of linear SVM at principal component numbers 5-25
plot_pca_accuracy(svm.LinearSVC(C = 1.0, random_state = 314), X_train_orig, X_test_orig, y_train, y_test)
```

```python
# Fit SVM using the optimal number of PC's (19)
pca = PCA(
    n_components = 19, 
    whiten = True, 
    random_state = 314
).fit(X_train_orig)

# Transform train/test data
X_train = pca.transform(X_train_orig)
X_test = pca.transform(X_test_orig)

# Fit model
svm_model = svm.LinearSVC(C = 1.0)
svm_model.fit(X_train, y_train)
print(svm_model.score(X_test, y_test))

# Confusion matrix
pred_svm = svm_model.predict(X_test)
confusion = confusion_matrix(y_test, pred_svm)
print("Confusion matrix:\n{}".format(confusion))
```

In the end, an SVM fitted on the principal components performed exceptionally well, with zero false negatives and only two false positives in the testing data.

```python
import os
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.neural_network import MLPRegressor
from sklearn.externals import joblib
```

```python
# Read data
LOCAL_DIR = os.path.join("data")
LOCAL_FILE = os.path.join(LOCAL_DIR,'housing.csv')
df = pd.read_csv(LOCAL_FILE)
df.info()
```

```python
# Remove island homes (imbalance)
df = df[df["ocean_proximity"] != 'ISLAND']

# Create dummy variables
df = pd.get_dummies(df)

# Preview data
df.head()
```

```python
# Median imputation
imputer = SimpleImputer(strategy = "median")
imputed = imputer.fit_transform(df)
df = pd.DataFrame(imputed, columns = df.columns)

# Preview changes
df.head()
```

```python
df['median_income'].hist()
```

```python
# Bucketize median income to stratify test/train/valid split
buckets = np.linspace(0,16,6)
df['median_income_cats'] = pd.cut(df['median_income'], buckets)
df['median_income_cats'].value_counts() / len(df)
```

```python
# Test/train split
X_train, X_test = train_test_split(
    df,
    test_size = 0.2,
    stratify = df['median_income_cats'],
    shuffle = True,
    random_state = 314
)
```

```python
# Train/validation split
X_train, X_valid = train_test_split(
    X_train,
    test_size = 0.15,
    stratify = X_train['median_income_cats'],
    shuffle = True,
    random_state = 314
)
```

```python
# Separate X & Y
y_train = X_train.loc[:, 'median_house_value'].values
y_valid = X_valid.loc[:, 'median_house_value'].values
y_test = X_test.loc[:, 'median_house_value'].values

X_train.drop('median_house_value', axis = 1, inplace = True)
X_valid.drop('median_house_value', axis = 1, inplace = True)
X_test.drop('median_house_value', axis = 1, inplace = True)

# Remove bucketized y as well
X_train.drop('median_income_cats', axis = 1, inplace = True)
X_valid.drop('median_income_cats', axis = 1, inplace = True)
X_test.drop('median_income_cats', axis = 1, inplace = True)
```

```python
# MinMax scale x
scalerStd = StandardScaler()
scalerStd.fit(X_train)

X_train_scaled = scalerStd.transform(X_train)
X_valid_scaled = scalerStd.transform(X_valid)
X_test_scaled = scalerStd.transform(X_test)
```

```python
# MinMax scale y
scalerStd_y = StandardScaler()
scalerStd_y.fit(y_train.reshape(-1, 1))

y_train_scaled = scalerStd_y.transform(y_train.reshape(-1, 1))
y_valid_scaled = scalerStd_y.transform(y_valid.reshape(-1, 1))
y_test_scaled = scalerStd_y.transform(y_test.reshape(-1, 1))
```

```python
# Reshape y
y_train_scaled = np.ravel(y_train_scaled)
y_valid_scaled = np.ravel(y_valid_scaled)
y_test_scaled = np.ravel(y_test_scaled)
```

```python
model = MLPRegressor()
model
```

```python
model = MLPRegressor(
    alpha = 0.001,
    hidden_layer_sizes = (200, 200)
).fit(
    X_train_scaled,
    y_train_scaled
)

preds_train = model.predict(X_train_scaled)
preds_valid = model.predict(X_valid_scaled)
print("Training R2 score: {:.2f}".format(r2_score(y_train_scaled, preds_train)))
print("validation R2 score: {:.2f}".format(r2_score(y_valid_scaled, preds_valid)))
```

```python
preds_test = model.predict(X_test_scaled)
print("Test R2 score: {:.2f}".format(r2_score(y_test_scaled, preds_test)))
```
